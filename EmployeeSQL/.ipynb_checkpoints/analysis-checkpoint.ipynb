```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/employee_db')
connection = engine.connect()
salaries_df = pd.read_sql_query('''SELECT s.salary, t.title FROM salaries AS s INNER JOIN employees AS e ON s.emp_no=e.emp_no INNER JOIN titles AS t ON t.title_id=e.emp_title_id;''', connection)
salaries_df['salary'] = salaries_df['salary'].astype('float')
```

```python
plt.hist(salaries_df['salary'], color='lightcoral')
plt.xlabel('Salary (USD)')
plt.ylabel('Number of Employees')
plt.title('Frequency of Salary Ranges at Pewlett Hackard')
plt.show()
```

```python
title_gb = salaries_df.groupby('title').mean()
title_gb.plot.bar(color='lightcoral')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.title('Average Salary by Job Title')
plt.show()
```
